## ***Image to Pencil Sketch with Python***

### **Importing required libraries**

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

### **Reading the image file**

```python
img=cv2.imread("img.png")
cv2_imshow(img)
```

```python
img.shape
```

### **Change size of the image for better view**

```python
resized = cv2.resize(img,(630,400),3)
cv2_imshow(resized)
```

### **Colour image to grey**

```python
gray = cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

### **Inverting Grey Image**

```python
inverted_gray = 255-gray
cv2_imshow(inverted_gray)
```

### **Apply Gaussian Blur**

```python
blur = cv2.GaussianBlur(inverted_gray,(21,21),0)
cv2_imshow(blur)
```

### **Inverted Blur Image**

```python
inverted_blur = 255-blur 
cv2_imshow(inverted_blur)
```

### **Blending Gray Image & Blur Image to generate image sketch**

```python
sketch = cv2.divide(gray,inverted_blur,scale=256.0)
cv2_imshow(sketch)
```
