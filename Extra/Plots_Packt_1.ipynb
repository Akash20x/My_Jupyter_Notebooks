```python
plt.figure(figsize=(10,8))
sns.countplot(df1['Payment Method'], hue=df1["Churn Value"])
plt.title("Distribution of Payment Method")
```

* Customers who made payment by Electronic check are mostly churn

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Contract', y='Monthly Charges', data=df1, hue='Churn Value')
plt.show()
```

* Customer with Month-to-month contract are having less monthly charge likely than One Year & Two Year contract as customer 
* .Customer with Month-to-month have less possibility of churn.
* Churn is high when monthly charges are high.

```python
plt.figure(figsize=(12,4))
df1['Churn Value'].value_counts().plot.pie(explode=[0,0.1], shadow=True, autopct='%1.1f%%')
```

* There is less percentage of customer to be churned.
